```python
import random
d1={'Plain cone':1.5,'Waffle cone':2,'Cup':1}
d2={'Vanilla':5,'Strawberry':5,'Chocolate':5,'Caramel':5,'Mint':5,'Rainbow':5,'Coffee':5,'Bubble gum':5}
d3={'Peanuts':0.75,'Caramel Sauce':0.5,'Rainbow Sprinkles':0.5,'Pecan':1,'Chocolate Sprinkles':0.5}
class Icecream:
    '''Welcome to our icecream shop'''
    def __init__(self,name):
        self.name=name
        self.l7=[]
        self.l8=[]
        self.cone=[]
        self.total=0
        self.l1=[]
        self.l2=[]
        for i in d3.keys():
            self.l1.append(i)
            self.l2.append(d3[i])
        self.l3=[]
        self.l4=[]
        for i in d2.keys():
            self.l3.append(i)
            self.l4.append(d2[i])
        self.l5=[]
        self.l6=[]
        for i in d1.keys():
            self.l5.append(i)
            self.l6.append(d1[i])
    def showtoppins(self):
        print("================================================================================")
        print("                               yummy Topins                                     ")
        print("================================================================================")
        for i in range(len(self.l1)):
            print("{:50}  {}".format(self.l1[i],self.l2[i]))
    def showflavour(self):
        print("===============================================================================")
        print("                              yummy Flavour                                    ")
        print("===============================================================================")
        for i in range(len(self.l3)):
            print("{:50}  {}".format(self.l3[i],self.l4[i]))
    def showcup(self):
        
        print("===============================================================================")
        print("                             yummy cuppy cup                                   ")
        print("===============================================================================")
        for i in range(len(self.l5)):
            print("{:50}  {}".format(self.l5[i],self.l6[i]))
```

```python
class ice(Icecream):
    def __init__(self,name):
        Icecream.__init__(self,name)
        self.total=0
    def toppin(self):
        self.total=0
        a=int(input('How many topins u want'))
        while(a>3):
            print("U can take only 3 topins")
            a=int(input("how many topins u want"))
        Icecream.showtoppins(self)
        i=0
        while(i<a):
            b=int(input("Enter choice from 1-6"))
            if(b>6):
                while(b>6):
                    print("Invalid choice")
                    b=int(input("Enter choice from 1-6"))
            for j in range(len(self.l2)):
                j+=1
                if(j==b):
                    j-=1
                    self.total=self.total+self.l2[j]
                    self.l7.append(self.l1[j])
                    self.l8.append(self.l2[j])
            i+=1
    def flv(self):
        a=int(input('How many flavours u want'))
        while(a>4):
            print("U can take only 4 flavours")
            a=int(input("how many flavour u want"))
        Icecream.showflavour(self)
        i=0
        while(i<a):
            b=int(input("Enter choice from 1-8"))
            if(b>8):
                while(b>8):
                    print("Invalid choice")
                    b=int(input("Enter choice from 1-8"))
            for j in range(len(self.l4)):
                j+=1
                if(j==b):
                    j-+1
                    self.total=self.total+self.l4[j]
                    self.l7.append(self.l3[j])
                    self.l8.append(self.l4[j])
            i+=1     
                    
    def cup1(self):
        Icecream.showcup(self)
        i=0
        b=int(input("Enter choice from 1-3"))
        if(b>3):
            while(b>3):
                print("Invalid choice")
                b=int(input("Enter choice from 1-3"))
        for i in range(len(self.l6)):
                i+=1
                if(i==b):
                    i-=1
                    self.total=self.total+self.l6[i]
                    self.l7.append(self.l5[i])
                    self.l8.append(self.l6[i])
    def calculate(self):
        sum=self.total
        print(sum)
        return(sum)
    def billno(self):
        self.billno=random.randint(1,10000)
    def tell(self,a):
        print("=========================================================================")
        print("                       Name of customer : ",self.name)
        print("=========================================================================")
        print("                       Bill no : ",self.billno)
        print("=========================================================================")
        for i in range(len(self.l7)):
            print("{:50} {}".format(self.l7[i],self.l8[i]))
        print("=========================================================================")
        print("{:50} {}".format('Total :',a))
        print("=========================================================================")
        
```

```python
while(1):
    a=input("Enter Name")
    p=ice(a)
    p.showtoppins()
    p.showflavour()
    p.showcup()
    b=int(input("How many ice cream u want"))
    i=0
    sum1=0
    while(i<b):
        p.toppin()
        p.flv()
        p.cup1()
        m=p.calculate()
        sum1+=m
        print(sum1)
        i+=1
    p.billno()
    p.tell(sum1)
        
```
